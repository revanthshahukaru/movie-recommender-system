```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculate item-item similarity matrix
item_similarity = cosine_similarity(df.pivot_table(index='id', columns='movieId', values='rating', fill_value=0).values)
```

```python
# Function to get movie ID based on movie name
def get_movie_id(movie_name):
    movie_id = df[df['original_title'] == movie_name]['id'].values[0]
    return movie_id
```

```python
# Function to get similar movies based on movie name
def get_similar_movies(movie_name, top_n=5):
    movie_id = get_movie_id(movie_name)
    
    # Get the similarity scores of the given movie with all other movies
    movie_similarity_scores = item_similarity[movie_id]
    
    # Get indices of top similar movies
    similar_movie_indices = movie_similarity_scores.argsort()[::-1][:top_n]
    
    # Get details of similar movies
    similar_movies = df[df['id'].isin(similar_movie_indices)]
    
    return similar_movies
```

```python
# Function to get movie ID based on movie name
def get_movie_id(movie_name):
    movie_id = df[df['original_title'] == movie_name]['id'].values[0]
    return movie_id
```

```python
# Example usage
input_movie_name = input("Enter a movie name: ")
similar_movies = get_similar_movies(input_movie_name, top_n=10)

# Display movie details
for index, movie in similar_movies.iterrows():
    print("Movie Name:", movie['original_title'])
    print("Original Language:", movie['original_language'])
    print("Genre String:", movie['genre_string'])
    print("----------------------------------")
```
